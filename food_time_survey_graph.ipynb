### Importing matplotlib.pyplot

```python
import matplotlib.pyplot as plt
```

### Making the line Graph (For Breakfast)

```python
vote = [9,5,4,2]
timeframe = ["7 to 9 AM","9 to 11 AM","The breakfast enters lunch time","Before 7 AM"]
plt.figure(figsize=(10,9))
plt.plot(timeframe, vote, color = 'orange', linestyle = '--', marker = 'o')
plt.xlabel('Timeframe')
plt.ylabel('Votes')
plt.title('Breakfast')
```

### Making the line Graph (For Lunch)

```python
vote = [13,6,1]
timeframe = ["1 PM to 3 PM","3 PM to 4 PM","The lunch time exceeds 4 PM"]
plt.figure(figsize=(10,9))
plt.plot(timeframe, vote, color = 'red', linestyle = '--', marker = 'o')
plt.xlabel('Timeframe')
plt.ylabel('Votes')
plt.title('Lunch')
```
